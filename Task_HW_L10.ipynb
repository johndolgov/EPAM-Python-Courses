### Какая страна выиграла большинство золотых медалей мужчинами в бадминтоне? Отсортировать результат в алфавитном порядке по имени игрока.

```python
badminton_table = olympic_table[olympic_table.Discipline == 'Badminton']
GoldMan_table = badminton_table[(badminton_table.Gender == 'Men') & (badminton_table.Medal == 'Gold')]
country_winner = GoldMan_table.NOC.value_counts().idxmax()
winners = GoldMan_table[GoldMan_table.NOC.str.contains(country_winner)]
print(f'The country that won most of the gold medals by men in badminton is {country_winner}' '\n')
print(f'Gold medalist of {country_winner} in alphabetic order ' '\n')
for i in winners.Athlete.sort_values():
    print(i)
```

### Какие три страны выйграли большинство медалей в последние годы (с 1984 по 2008)?

```python
olympic_table_1984_2008 = olympic_table[olympic_table.Edition >= 1984]
max_olympic = olympic_table_1984_2008.NOC.value_counts()
print('This three countries won most of the gold medals' '\n')
for i in range(3):
    print(max_olympic.index[i])
```

## Task #L10.4

### Постройте график числа золотых медалей выйгранных США мужчинами и женщинами в атлетике?

```python
USA_Athletes = olympic_table[(olympic_table.NOC == 'USA') & (olympic_table.Discipline == 'Athletics') & (olympic_table.Medal == 'Gold')]
USA_Athletes_Group = USA_Athletes.groupby(['Edition', 'Gender']).size()
Unstacked_Atheletes = USA_Athletes_Group.unstack('Gender')
Unstacked_Atheletes.plot(kind = 'bar')
plt.ylabel('Quantity of the Medals');
```

### Постройте график 5 атлетов которые выйграли большинство золотых медалей. 

```python
Best_Athletes = olympic_table[olympic_table.Medal == 'Gold'].groupby('Athlete').size()
Best_Athletes.sort_values(ascending = False).head(5).plot(kind = 'bar');
plt.ylabel('Quantity of the medals');
```
